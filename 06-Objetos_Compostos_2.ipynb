## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
genes = ["ESR1", "ESR2", "PGR", "ERBB2", "BRCA1", "BRCA2"]

matriz = []

for gene in genes:
    vetor = [
        len(gene),
        gene[:2],
        gene.lower().count("a"),
        "hu" in gene.lower()
    ]
    matriz.append(vetor)

print(matriz)
```

## 3 – Copie e converta o objeto instanciado no exercício 2 para um NumPy array. Recupere o array.

```python
import numpy as np

array = np.array(matriz)
print(array)
```

## 5 – Verifique se todos os nomes dos genes tem ou não os caracteres 'hu'.

```python
resultado = []

for gene in genes:
    resultado.append("hu" in gene.lower())

print(resultado)
```

## 7 – Converta o objeto instanciado no exercício 10 da aula anterior para uma DataFrame Pandas. Nesta DataFrame, os index devem ser os nomes dos genes e cada coluna deve ter um nome que indique seu conteúdo. Recupere a DataFrame.

```python
import pandas as pd

dicionario = {genes[i]: matriz[i] for i in range(len(genes))}
df = pd.DataFrame.from_dict(
    dicionario,
    orient="index",
    columns=["num_caracteres", "dois_primeiros", "qtd_a", "tem_hu"]
)

print(df)
```

## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
import pandas as pd

df = pd.read_csv("data.csv")
num_entradas = df.shape[0]
classes = list(df.dtypes)
num_por_classe = df.dtypes.value_counts().tolist()
colunas = df.iloc[:, [0, 3, 8]]
medias = colunas.mean().tolist()
desvios = colunas.std().tolist()
q3 = colunas.quantile(0.75).tolist()

resultado = {
    "num_entradas": num_entradas,
    "classes": classes,
    "num_por_classe": num_por_classe,
    "medias": medias,
    "desvios": desvios,
    "q3": q3
}

print(resultado)
```
