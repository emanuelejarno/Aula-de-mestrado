## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Gerei uma lista de números de 0 a 4999 e usei o resto da divisão por 2 para separar pares e ímpares. Depois, contei quantos há em cada grupo.

## 02 – Desenhe o fluxograma da resposta:

## 03 – Escreva o programa em Python:

```python
lista = list(range(5000))

pares = len([x for x in lista if x % 2 == 0])
impares = len([x for x in lista if x % 2 != 0])

print(pares)
print(impares)
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
lista = list(range(5000))

pares = sum(1 for x in lista if x % 2 == 0)
impares = len(lista) - pares

print(pares)
print(impares)
```

## 06 – Desenhe o fluxograma da resposta:

## 07 – Escreva o programa em Python:

```python
import pandas as pd

df = pd.read_csv("dataframel6e7.csv", index_col="Unnamed: 0")

pares = df[df["num_caracteres"] % 2 == 0]["num_caracteres"]
impares = df[df["num_caracteres"] % 2 != 0]["num_caracteres"]

print("Números pares:")
print(pares)

print("Quantidade de ímpares:")
print(len(impares))
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
import pandas as pd

pares = df.loc[df["num_caracteres"] % 2 == 0, "num_caracteres"]
quantidade_impares = (df["num_caracteres"] % 2 != 0).sum()

print("Números pares:")
print(pares)

print("Quantidade de ímpares:")
print(quantidade_impares)
```
