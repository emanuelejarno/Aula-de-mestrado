## Responda a pergunta do exercício 2 da aula 10, usando funções:<br>
Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R.: O código foi organizado em funções para facilitar a leitura e evitar deixar tudo misturado. Primeiro, uma função cria um array com 5 valores aleatórios, seguindo uma distribuição normal com média 10 e desvio padrão 3. Isso significa que os números gerados tendem a ficar próximos de 10, mas podem variar para mais ou para menos.

Depois, outra função recebe esse array e compara cada valor com 10. Quando um valor é maior que 10, ele é selecionado e armazenado em um novo conjunto de valores. Em seguida, o código conta quantos valores passaram nessa condição.

No final, o programa mostra três informações: o array original gerado, a quantidade de valores superiores a 10 e quais são esses valores. 

## 03 – Desenhe o fluxograma da resposta:

## 04 – Escreva o programa em Python:

```python
import numpy as np

def criar_array():
    return np.random.normal(loc=10, scale=3, size=5)

def valores_maiores_que_10(array):
    valores = array[array > 10]
    quantidade = len(valores)
    return quantidade, valores

array = criar_array()
quantidade, valores = valores_maiores_que_10(array)

print(array)
print("Quantidade de valores superiores a 10:", quantidade)
print("Valores superiores a 10:", valores)
```

## Desenvolva um programa de computador que retorne a média de um conjunto de valores.

## 06 – Descreva o racional da sua resposta:

R.: O código pega uma lista de números, soma todos eles e divide pela quantidade de elementos da lista. Esse cálculo gera a média, que depois é mostrada na tela.

## 07 – Desenhe o fluxograma da resposta:

## 07 – Escreva o programa em Python:

```python
def calcular_media(valores):
    soma = sum(valores)
    quantidade = len(valores)
    media = soma / quantidade
    return media

valores = [10, 8, 7, 9, 6]

resultado = calcular_media(valores)

print("A média dos valores é:", resultado)
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_objetos(a, b):
    print("Objeto 1:")
    print("Valor:", a)
    print("Classe:", type(a))

    print("Objeto 2:")
    print("Valor:", b)
    print("Classe:", type(b))

    resultado = a / b

    return resultado


objeto1 = 10
objeto2 = 2

divisao = dividir_objetos(objeto1, objeto2)

print("Resultado da divisão:", divisao)
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
import numpy as np

def dividir_objetos_compostos(objeto1, objeto2):
    print("Objeto 1:")
    print("Valor:", objeto1)
    print("Classe:", type(objeto1))

    print("Objeto 2:")
    print("Valor:", objeto2)
    print("Classe:", type(objeto2))

    array1 = np.array(objeto1)
    array2 = np.array(objeto2)

    resultado = array1 / array2

    return resultado


objeto1 = [10, 20, 30]
objeto2 = [2, 4, 5]

divisao = dividir_objetos_compostos(objeto1, objeto2)

print("Resultado da divisão:", divisao)
```
