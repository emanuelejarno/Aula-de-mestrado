## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t
invertido = not resultado

print(resultado)
print(invertido)
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

resultado = p or q or r or s or t or u
invertido = not resultado

print(resultado)
print(invertido)
```

## 03 – No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas. Importe aqui o mesmo objeto e arredonde e edite o index como realizado no exercício 3 da aula 8. Usando operadores responda, a segunda linha é igual a terceira ou a terceira linha é igual a quarta?

```python
import pandas as pd

df = pd.read_csv("dataframel7e10.csv", index_col="Unnamed: 0")
df = df.round(3)

df.index = ["linha1", "linha2", "linha3", "linha4", "linha5", "media_colunas"]

resultado = (df.iloc[1].equals(df.iloc[2])) or (df.iloc[2].equals(df.iloc[3]))

print(resultado)
```

## 04 – Usando operadores verifique se todos os valores contidos na primeira linha são maiores ou iguais a 10.

```python
resultado = (df.iloc[0] >= 10).all()
print(resultado)
```

## 05 – Usando operadores verifique se todos os valores contidos na terceira coluna estão entre 0 inclusive e 40.

```python
resultado = ((df.iloc[:, 2] >= 0) & (df.iloc[:, 2] <= 40)).all()
print(resultado)
```

## 06 – Usando operadores verifique se algum dos valores contidos na última linha está acima de 10.

```python
resultado = (df.iloc[-1] > 10).any()
print(resultado)
```

## 07 – Recupere a DataFrame contendo apenas valores que estejam acima de 5 e abaixo de 7.

```python
df_filtrado = df[(df > 5) & (df < 7)]
print(df_filtrado)
```

## 08 – Recupere a DataFrame contendo apenas valores que estejam abaixo de 5 inclusive ou acima de 7.

```python
df_filtrado = df[(df <= 5) | (df > 7)]
print(df_filtrado)
```
